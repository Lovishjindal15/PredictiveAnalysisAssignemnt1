```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
from scipy.stats import norm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

def solve_pdf_assignment(roll_number):

    r = roll_number
    ar = 0.05 * (r % 7)
    br = 0.3 * (r % 5 + 1)
    print('ar: ',ar)
    print('br: ',br)



    df = pd.read_csv('/content/data.csv', encoding='latin-1')
    print(f"Total records in dataset: {len(df)}")


    x = df['no2'].dropna()
    x = pd.to_numeric(x, errors='coerce').dropna()
    x = x[x > 0]
    x = x.values



    #  Transform x to z using the transformation function


    z = x + ar * np.sin(br * x)




    #  Learn PDF parameters using Maximum Likelihood Estimation
    # p(z) = c * exp(-λ(z-μ)²)
    # This is a Gaussian PDF where:
    #   μ = mean
    #   λ = 1/(2sigma²)
    #   c = √(lambda/pie)





    mu_mle = np.mean(z)
    sigma2_mle = np.var(z)
    lambda_mle = 1 / (2 * sigma2_mle)
    c_mle = np.sqrt(lambda_mle / np.pi)







    # FINAL RESULTS




    print("\n Parameters:")
    print(f"λ = {lambda_mle}")
    print(f"μ = {mu_mle}")
    print(f"c = {c_mle}")



if __name__ == "__main__":

    ROLL_NUMBER = 102303950

    results = solve_pdf_assignment(ROLL_NUMBER)
```
